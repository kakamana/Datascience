```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
country_data=pd.read_csv("countries-of-the-world.csv")
country_data.head()
```

```python
gdp=country_data["GDP ($ per capita)"].tolist()
phones=country_data["Phones (per 1000)"].tolist()
percent_literate=country_data["Literacy (%)"].tolist()
region=country_data["Region"].tolist()
```

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp,y=phones)
plt.show()
```

```python
sns.scatterplot(x=gdp, y=percent_literate)

# Show plot
plt.show()
```

Count plot

```python
# Create count plot with region on the y-axis
sns.countplot(y=region)

# Show plot
plt.show()
```

```python
younPeople=pd.read_csv("young-people-survey-responses.csv")
younPeople.head()
```

```python
# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Spiders",data=younPeople)

# Display the plot
plt.show()
```

## Customizing scatter plot

```python
mpg=pd.read_csv("mpg.csv")
mpg.head()
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",hue="cylinders")

# Show plot
plt.show()
```

Cars with higher horsepower tend to get a lower number of miles per gallon. They also tend to have a higher number of cylinders.

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x="acceleration",y="mpg",data=mpg,style="origin",hue="origin")

# Show plot
plt.show()
```

Cars from the USA tend to accelerate more quickly and get lower miles per gallon compared to cars from Europe and Japan.

## Line plot in relational plotting

```python
# Create line plot
sns.relplot(x="model_year",y="mpg",data=mpg,kind="line")

# Show plot
plt.show()
```

Replacing confidence internal with standard deviation

```python
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line",ci="sd")

# Show plot
plt.show()
```

## Plotting subgroups in line plots

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None,style="origin",hue="origin",markers=True,dashes=False)

# Show plot
plt.show()
```

## Customizing Seaborn Plots
### Adding a title to a FacetGrid object

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```
